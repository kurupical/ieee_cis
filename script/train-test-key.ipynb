```python
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
df_train = pd.read_feather("../data/baseline/train/baseline.feather")
df_test = pd.read_feather("../data/baseline/test/baseline.feather")
```

```python
df_train["TEMP__uid2+DT"].isin(df_test["TEMP__uid2+DT"]).sum()
```

```python
df_train_isfraud = df_train[["TEMP__uid2+DT", "isFraud"]].groupby("TEMP__uid2+DT").agg({"isFraud": ["count", "mean"]}).reset_index()
df_train_isfraud.columns = ["TEMP__uid2+DT", "count_train", "mean"]
```

```python
df_test_isfraud = df_test["TEMP__uid2+DT"].value_counts().reset_index()
df_test_isfraud.columns = ["TEMP__uid2+DT", "count_test"]
```

```python
pd.merge(df_test_isfraud, df_train_isfraud, how="left").to_csv("isfraud_count_merge.csv", index=False)
```

```python
df_train_isfraud[df_train["TEMP__uid2+DT"].isin(df_test["TEMP__uid2+DT"])].to_csv("TEMP__uid2+DT_train.csv", index=False)
```

```python
df_merge = pd.merge(df_test[["TEMP__uid2+DT", "TransactionID", "D1", "TEMP__DT_D"]], df_train_isfraud, how="left")
```

```python
df_merge.sort_values("mean", ascending=False)
```
