```python
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
df_train = pd.read_feather("../data/baseline/train/baseline.feather").query("D1 > 0").reset_index()
df_test = pd.read_feather("../data/baseline/test/baseline.feather")
```

```python
df_train["TEMP__uid2+DT2"].isin(df_test["TEMP__uid2+DT2"]).sum()
```

```python
df_train_isfraud = df_train[["TEMP__uid2+DT2", "isFraud"]].groupby("TEMP__uid2+DT2").agg({"isFraud": ["count", "mean"]}).reset_index()
df_train_isfraud.columns = ["TEMP__uid2+DT2", "count_train", "mean"]
```

```python
df_test_isfraud = df_test["TEMP__uid2+DT2"].value_counts().reset_index()
df_test_isfraud.columns = ["TEMP__uid2+DT2", "count_test"]
```

```python
pd.merge(df_test_isfraud, df_train_isfraud, how="left").to_csv("isfraud_count_merge_DT2.csv", index=False)
```

```python
df_train_isfraud = pd.merge(df_test_isfraud, df_train_isfraud, how="left")
```

```python
df_agg = df_train[["TEMP__uid2+DT2", "D4"]].groupby("TEMP__uid2+DT2").agg({"D4": ["min", "max"]}).reset_index()
df_agg.columns = ["TEMP__uid2+DT2", "min_D4", "max_D4"]
```

```python
df_agg["diff_D4maxmin"] = df_agg["max_D4"] - df_agg["min_D4"]
```

```python
df_agg = pd.merge(df_train[["TEMP__uid2+DT2", "isFraud"]], df_agg, how="left", on="TEMP__uid2+DT2")
```

```python
df_agg = df_agg.groupby(["TEMP__uid2+DT2", "diff_D4maxmin"]).agg({"isFraud": ["mean", "count"]}).reset_index()
df_agg.columns = ["TEMP__uid2+DT2", "diff_D4maxmin", "mean", "count"]
```
