```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_feather("../data/original/train_all.feather")
df_test = pd.read_feather("../data/original/test_all.feather")
```

```python
df_train = df_train.groupby(["card1", "card2", "card3"])["TransactionID"].count()
df_test = df_test.groupby(["card1", "card2", "card3"])["TransactionID"].count()
```

```python
df_all = pd.concat([df_train, df_test], axis=1).fillna(0)
df_all.columns = ["train", "test"]
df_all["train/test"] = df_all["train"] / df_all["test"]
```

```python
df_all["train/test"] = df_all["train/test"].replace(np.inf, 0)
```

```python
df_all.sort_values("train/test", ascending=False)
```
