```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_feather("../data/original/train_transaction.feather")
```

```python
for i in range(1, 1+6):
    w_df = df.groupby("card{}".format(i)).agg({"isFraud": ["count", "mean"]})
    w_df.columns = ["count", "mean"]
    print(w_df.sort_values("count", ascending=False))
    w_df.sort_values("count", ascending=False).to_csv("cardcol/agg/card{}.csv".format(i))
    plt.plot(w_df["mean"].values)
    plt.show()
```
