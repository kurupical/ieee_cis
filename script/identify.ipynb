```python
top_idx = len(df_all) - 1
identify_id = 0

w_df = df_all.iloc[top_idx]
df_all.at[w_df.name,"Identify_ID"] = identify_id
w_day = w_df["TransactionDay"]

while True:
    w_df = w_df.fillna(-100)

    is_chain_end = False
    # print(w_df[["D1", "D2", "D3", "D4", "D5"]].isnull().sum())
    if len(w_df) == 0:
        is_chain_end = True
    elif w_df[["D1", "D2", "D3", "D4", "D5"]].isnull().sum() == 5:
        is_chain_end = True
    elif w_df[["D1", "D2", "D3", "D4", "D5"]].std() == 0:
        is_chain_end = True
    elif w_df["Identify_ID"] != "":
        is_chain_end = True
    elif w_df["TransactionDay"] - w_df["D5"] < 0:
        is_chain_end = True
    elif np.isnan(w_df["D5"]):
        is_chain_end = True
    
    if len(w_df) > 0:
        df_all.at[w_df.name,"Identify_ID"] = "ID_{}".format(identify_id)

    if is_chain_end:
        identify_id += 1
        top_idx = top_idx - 1
        w_df = df_all.iloc[top_idx]
        
        if identify_id % 100 == 0:
            print("ID = {} / Identified: {}".format(identify_id, len(df_all.query("Identify_ID != ''"))))
        if top_idx >= len(df_all):
            break
    else:
        query = "TransactionDay >= {}-1 and TransactionDay <= {}+1".format(w_day-w_df["D5"], w_day-w_df["D5"])
        query += " and "
        query += "card1 == '{}' and card2 == '{}' and card3 == '{}'".format(w_df["card1"], w_df["card2"], w_df["card3"])
        query += " and "
        query += "(D10 >= {}-1 and D10 <= {}+1".format(w_df["D10"]-w_df["D5"], w_df["D10"]-w_df["D5"])
        query += " or "
        query += " D11 >= {}-1 and D11 <= {}+1".format(w_df["D11"]-w_df["D5"], w_df["D11"]-w_df["D5"])
        query += " or "
        query += " D15 >= {}-1 and D15 <= {}+1".format(w_df["D15"]-w_df["D5"], w_df["D15"]-w_df["D5"])
        query += ")"
        # print(query)
        w_df = df_all.query(query)

        w_df = w_df.sort_values("D5")

        for i in range(len(w_df)):
            if w_df.iloc[i]["D5"] == 0:
                df_all.at[w_df.iloc[i].name,"Identify_ID"] = "ID_{}".format(identify_id)
            else:
                w_df = w_df.iloc[i]
                break
            if i+1 == len(w_df):
                w_df = w_df.iloc[i]
```
